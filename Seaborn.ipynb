Data Visualization of [this dataset](https://www.kaggle.com/datasets/piterfm/2022-ukraine-russian-war)

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('Setup Complete')
```

```python
# Set seaborn style
# sns.set_style('darkgrid')
# sns.set_style('whitegrid')
sns.set_style('white')
# sns.set_style('dark')
```

```python
my_data = pd.read_csv('/content/russia_losses_equipment.csv')
del my_data['date']
my_data
```

```python
plt.figure(figsize=(16,9))
sns.lineplot(data=my_data)
```

```python
plt.figure(figsize=(16,10))
sns.barplot(x=my_data.index, y=my_data['aircraft'])
```

```python
plt.figure(figsize=(14,14))
# sns.heatmap(data=my_data, annot=True)
sns.heatmap(data=my_data)
```

```python
plt.figure(figsize=(16,8))
sns.scatterplot(x=my_data['day'], y=my_data['helicopter'])
```

```python
plt.figure(figsize=(16,8))
sns.scatterplot(x=my_data['day'], y=my_data['tank'], hue=my_data['fuel tank'])
```

```python
plt.figure(figsize=(16,8))
# sns.histplot(data=my_data['military auto'], kde=False)
sns.histplot(data=my_data['military auto'], kde=True)
```

```python
plt.figure(figsize=(16,8))
sns.histplot(data=my_data, multiple="stack")
```

```python
# When both x and y are assigned, a bivariate histogram is computed and shown as a heatmap
plt.figure(figsize=(16,8))
sns.histplot(data=my_data, x="day", y='tank')
```

```python
plt.figure(figsize=(16,8))
sns.regplot(x=my_data['day'], y=my_data['naval ship'])
# sns.regplot(x=my_data['day'], y=my_data['drone'])
```

```python
plt.figure(figsize=(16,8))
sns.lmplot(x='helicopter', y='aircraft', hue='anti-aircraft warfare', data=my_data)
```

```python
plt.figure(figsize=(16,8))
sns.swarmplot(x=my_data['tank'], y=my_data['fuel tank'])
```

```python
plt.figure(figsize=(16,8))
sns.kdeplot(data=my_data['APC'], shade=True)
```

```python
plt.figure(figsize=(16,8))
sns.jointplot(x=my_data['aircraft'], y=my_data['anti-aircraft warfare'], kind='kde')
```
