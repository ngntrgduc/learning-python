```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = Image.open(r"pikachu.jpg") 
print(image)
```

```python
arr = np.array(image)
print(arr)
```

```python
print(arr.ndim)
print(arr.shape)
print(arr.size)
```

```python
def show(img):
    plt.imshow(img)
    pass
show(image)
```

```python
img_R, img_G, img_B = arr.copy(), arr.copy(), arr.copy()
img_R[:, :, (1, 2)] = 0
img_G[:, :, (0, 2)] = 0
img_B[:, :, (0, 1)] = 0
img_RGB = np.concatenate((img_R, img_G, img_B), axis=1)
show(img_RGB)
```
