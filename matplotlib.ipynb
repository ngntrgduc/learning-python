```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 2, 500)
y = x**2
plt.plot(x, y, label='$y = x^2$')
plt.title("Square function")
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.grid(True)
# plt.gca().set_aspect('equal')
(px, py) = 0, 0
plt.plot(px, py, "ro")
plt.text(px + 1.5, py + 2, "$y = x^2$", fontsize=10, color='r',
         ha="left") # ha = horizontalalignment 
# plt.text(px, py + 0.5, "Local minimum", ha="center", weight="heavy")
plt.annotate("Local minimum", xy=(px, py), xytext=(px, py + 0.5), 
             ha="center", fontsize=10,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=0.2")) 

# The default legend behavior for axes attempts to find the location that covers
# the fewest data points (loc='best'). This can be a very expensive computation 
# if there are lots of data points => provide a specific location.
plt.legend(loc='lower right', frameon=False) # = plt.legend(loc=4)

# plt.minorticks_off() # disable minorticks
plt.xticks([]) 
plt.yticks([]) # xticks and yticks disable plt.grid(True)

# plt.tick_params(length=5, width=2, grid_alpha=0.1)
# plt.margins(0.2)
# sns.despine() # get rid of top spine and right spine

# plt.savefig("square_function.png", transparent=True)

plt.show()
```

```python
colors = ['C0', 'deepskyblue', 'hotpink', 'red',  'C3']
# colors = ['C0', 'C3']
# colors = ['deepskyblue', 'hotpink', 'red']
# colors = [f'C{i}' for i in range(10)]

for color in colors:
    line = np.random.randint(30, size=5)
    plt.plot(line, label=f'{color}', color=color)

plt.legend(bbox_to_anchor=(1, 0.5), frameon=False)
plt.axis('off')
plt.box(False)
plt.show()
```

```python
x = [-3, -2, 0, 5]
y = [1, 2, 3, 4]
plt.subplot(1, 2, 1)
plt.title('Normal plot')
plt.plot(x, y)

plt.subplot(1, 2, 2)
plt.title('Plot with axis')
plt.axis([-4, 6, 0, 5]) # [xmin, xmax, ymin, ymax]
plt.plot(x, y)

plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
    
x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.show()
```

```python
colormaps = [['viridis', 'plasma', 'inferno', 'turbo'],
             ['Blues', 'PuRd', 'RdPu', 'GnBu'],
             ['binary', 'spring', 'winter', 'cool', 'hot']]
for row, colors_map in enumerate(colormaps):
    plt.figure(figsize=(15, 15))
    for col, color_map in enumerate(colors_map):
        plt.subplot(row+1, len(colors_map), col+1)
        plt.axis('off')

        im = plt.imshow(np.arange(10*10).reshape(10, 10), cmap=color_map)
        plt.colorbar(im, fraction=0.045, pad=0.04).set_ticks([])

    plt.show()
```
