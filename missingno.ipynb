```python
import missingno as msno
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
collisions = pd.read_csv("https://raw.githubusercontent.com/ResidentMario/missingno-data/master/nyc_collision_factors.csv")
collisions.sample(5)
```

```python
# msno.matrix(collisions.sample(250))
# msno.matrix(collisions.sample(250), sparkline=False)
msno.matrix(collisions, sparkline=False)
```

```python
# msno.bar(collisions.sample(1000))
msno.bar(collisions)
```

```python
msno.heatmap(collisions)
```

```python
msno.dendrogram(collisions, orientation='right')
```
